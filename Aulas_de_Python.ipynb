```python
# Criação de um dataframe de teste para exemplo
import numpy as np

df_salarios = pd.DataFrame({
    'nome': ['pedro', 'carlos', 'roberto', 'bruna', 'sara'],
    'salario': [4000 , np.nan, 5000, np.nan, 100000]
})
# calcula a média e substitui o nulo por ela / arredonda valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

''' calcula mediana e substitui os nnulos pela mediana
'''
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['segunda','terça','quarta','quinta','sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

# ffill preenche o nulo com o número anterior
df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['segunda','terça','quarta','quinta','sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

# bfill preenche o nulo com o próximo número
df_temperaturas['preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
```

```python
df_cidade = pd.DataFrame({
    'nome': ['Pedro', 'Carlos', 'Roberto', 'Bruna', 'Sara'],
    'cidade': ['São Paulo', np.nan, 'Curítiba', np.nan, 'Belém']
})

df_cidade['cidade_preenchida'] = df_cidade['cidade'].fillna("não informado")
df_cidade
```

```python
import pycountry
```

```python
df_cientista_dados = df_limpo[df_limpo['job_title'] == 'Data Scientist']

salario_por_pais = df_cientista_dados.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).reset_index()
```

```python
# Function to get country name from ISO-3 code
def get_country_name(iso_code):
    try:
        return pycountry.countries.get(alpha_3=iso_code).name
    except AttributeError:
        return None

# Apply the function to create a new column with full country names
salario_por_pais['full_country_name'] = salario_por_pais['company_location'].apply(get_country_name)

# Display the updated DataFrame to verify
display(salario_por_pais.head())
```

```python
fig_mapa_pycountry = px.choropleth(
    salario_por_pais.dropna(subset=['full_country_name']), # Drop rows where country name couldn't be found
    locations='company_location',
    locationmode='ISO-3',
    color='salary_in_usd',
    hover_name='full_country_name', # Use the full country name for hover
    color_continuous_scale=px.colors.sequential.Plasma,
    title='Salário Médio de Cientista de Dados por País (Mapa Mundi com Nomes Completos)'
)

fig_mapa_pycountry.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular'
    ),
    annotations = [dict(
        x=0.5, y=0,
        xref='paper', yref='paper',
        text='Fonte: company_location usa códigos ISO-3. Alguns países podem não ser exibidos devido à falta de dados ou incompatibilidade de códigos.',
        showarrow = False
    )]
)

fig_mapa_pycountry.show()
```
